### Arrays

Un **array** es una estructura de datos que consiste en una conexion de elementos (valores o variables), cada uno identificado por al menos un indice o clave. Un array se almacena de modo que la posicion de cada elemento se pueda calcular a partir de su tupla de indice mediante su formula matematica. El tipo mas simple del array es el array lineal, tambien llamado array unidimencional

En Numpy:

* Cada dimension se denomina **axis**.
* El numero de dimenciones se denomina **rank**.
* La lista de dimenciones con su correspondiente longitud se denomina **shape**.
* El numero total de elementos (que es la multiplicacion de la longitud de las dimensiones) se denomina  **size** 

```python
# Array cuyos valores son todos 0 
a = np.zeros((2, 4))
```

```python
a
```

**a** es un array:

* Con 2 **axis**, la primera de longitud 2 y el segundo de longitud 4.
* Con un **rank** igual a 2.
* Con un **shape** igual (2,4).
* Con un **size** igual a 8.

### Acceso a los elementos de un Array

#### Array Unidimensional

```python
#Crear un array unidimencional 
array_uni = np.array([1, 3, 5, 7, 9, 11])
print("Shape: ", array_uni.shape)
print ("Array_uni: ", array_uni)
```

```python
# Acceder al 5to elemento del array 
array_uni[4]
```

```python
# Acceder al 3er y 4to elemto del array
array_uni[2: 4]
```

```python
#Acceder al elemento 0 3 y 5 del array
array_uni[[0, 3, 5]]
```

#### Array Multidimensional

```python
#Crear in array multidimencional
array_multi = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("Shape: ", array_multi.shape)
print("Array_multi: \n", array_multi)
```

```python
#Acceder al cuarto elemento del array
array_multi[0, 3]
```

```python
#Acceder a la segunda fila del array
array_multi[1, :]
```

```python
#Acceder al tercer elemento de las 2 primeras filas del array
array_multi[0: 2, 2]
```

#### Modificar un Array

```python
#Crear un array unidimencional e inicializarlo con el rango de elementos del 0 al 27
array1 = np.arange(28)
print("Shape: ", array1.shape)
print ("Array1: \n", array1)
```

```python
#Cambiar las dimensiones del array y sus longitudes
array1.shape = (7, 4 )
print("Shape: ", array1.shape)
print ("Array1: \n", array1)
```

```python
# Ejemplo anterior devuelve un nuevo array que apunta a los mismos datos
# NOTA: LAS MODIFICACIONES DE UN ARRAY MODIFICAN EL OTRO ARRAY
array2 = array1.reshape(4, 7)
print("Shape: ", array2.shape)
print ("Array2: \n", array2)
print ("Array1: \n", array1)
```

```python
#Modificar el array devuelto
array2[0, 3] = 20
print ("Array1: \n", array1)
print ("Array2: \n", array2)
```

```python
#Regresar el array devolviendo un nuevo array de una sola dimension 
#NOTA: EL NUEVO ARRAY APUNTA A LOS MISMOS DATOS
print ("Array1: ", array1.ravel())
```

### Operaciones Aritmeticas con Arrays

```python
import numpy as np
```

```python
array1 = np.arange(2, 18, 2)
array2 = np.arange(8)
print("Array 1:", array1)
print("Array 2:", array2)
```

```python
#SUMA
print(array1 + array2)
```

```python
#RESTA
print(array1 - array2)
```

```python
#Multiplicacion
print(array1 * array2)
```

##Broadcasting
si se aplican operaciones aritmeticas sobre arrays que no tengan las misma forma shape Numpy aplica una propiedad que se denomina BROADCASTING

```python
#Creacion de dos arrays
array1 = np.arange(5)
array2 = np.array([3])
print("Shape Array 1:", array1.shape)
print("Array 1:", array1)
print("\n")
print("Shape Array 2:", array2.shape)
print("Array 2:", array2)
```

```python
#Suma de dos arreglos
array1 + array2
```

```python
# Crear un arreglo Multidimencional y Unidimencional 
array1 = np.arange(6)
array1.shape = (2, 3)
array2 = np.arange(6, 18, 4)
print("Shape Array 1:", array1.shape)
print("Array 1:", array1)
print("")
print("Shape Array 2:", array2.shape)
print("Array 2:", array2)
```

```python
# Suma de ambos arrays 
array1 + array2
```
